```python
fig=plt.figure()
ax=fig.add_subplot(1,1,1)
rect=plt.Rectangle((0.2,0.75),0.4,0.05,color='y',alpha=0.3)
circ=plt.Circle((0.7,0.7),0.10,color='b',alpha=0.3)
pgon=plt.Polygon([[0.15,0.05],[0.35,0.4],[0.2,0.6]],color='g',alpha=0.9)
ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
#from pandas import DataFrame
data = pd.read_csv('plotdata.txt'," ")
df=pd.DataFrame(data)
df.columns = ['height','weight']
plt.subplot(121)# rownum/colnum/locindex #1代表行，2代表列有2个图，1代表此时绘制第一个图。
plt.plot(df)
plt.title('fat figure')
plt.legend(["height", "weight"])
data = pd.read_csv('plotdata2.txt'," "，header=None)#加header=None就不会少一个数据
df=pd.DataFrame(data)
df.columns = ['python','php','java']
plt.subplot(122)
plt.plot(df)
plt.title('language figure')
plt.legend(['python','php','java'])
```
