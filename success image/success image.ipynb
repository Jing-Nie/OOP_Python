```python
import jieba
from wordcloud import WordCloud
import matplotlib.pyplot as plt
f = open('Mandela.txt','r',encoding='UTF-8').read()
#生成词云
wordcloud=WordCloud(background_color="white",font_path="STKAITI.TTF",width=1000,height=1000,margin=2 ).generate(f)
wordcloud.generate(" ".join(jieba.lcut(f)))
#显示词云图片
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
wordcloud.to_file("image.png")#“非国大”是“非洲人国民大会”的缩写
```

```python
import jieba
from wordcloud import WordCloud,ImageColorGenerator
import matplotlib.pyplot as plt
from scipy.misc import imread
f = open('Mandela.txt','r',encoding='UTF-8').read()
#读入背景图片
bg_pic = imread('black1.png')
#生成词云
wordcloud=WordCloud(mask=bg_pic,background_color="white",font_path="STKAITI.TTF",scale=1.5).generate(f)
wordcloud.generate(" ".join(jieba.lcut(f)))
image_colors = ImageColorGenerator(bg_pic)
#显示词云图片
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
wordcloud.to_file("image2.png")
```
